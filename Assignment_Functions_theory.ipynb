### **Question 2. Explain the concept of function arguments and parameters in Python.**



In Python, function arguments and parameters are related concepts, but they refer to different parts of a function. Here's an explanation of both:

1. **Function Parameters**: These are the names listed in the function definition. They act as placeholders for the values that will be passed to the function when it is called.

2. **Function Arguments**: These are the actual values or data you pass to the function when calling it. The values are assigned to the corresponding parameters defined in the function.

In summary, parameters are part of the function's definition, while arguments are part of the function's invocation.

```python
#example to define a function with parameters
def greet(name, age):
    print(f"Hello {name}, you are {age} years old.")

#Calling the function with arguments
greet("Azlan", 27)
```

Explanation of the given example.

**Parameters:**

In the function definition def greet(name, age):, name and age are parameters. They are like placeholders waiting for actual values when the function is called.

**Arguments:**

When you call the function with greet("Azlan", 27), "Azlan" and 27 are arguments. These are the actual values passed to the function, which are assigned to the name and age parameters.

### **Question 6.  Explain the concept of generators in Python and how they are defined?**

In Python, generators are a special type of function that allows us to create iterators in a more concise and efficient way. They use the "yield" keyword instead of "return" to produce a sequence of values one at a time, allowing for memory-efficient iteration when working with large datasets.

When a function contains a "yield" statement, it becomes a generator function.
Each time "yield" is called, the function produces a value, and the function's state is saved so it can resume from where it left off the next time it is called.

```python
#example of generator function.

def my_generator(n):
  value = 0      #generator function that yields numbers from 0 to n.
  while value < n:
    yield value
    value += 1 #increasing the value
for i in my_generator(7):
  print(i)
```

```python
#example for filtering even number
def even_numbers(xyz):
    for num in xyz:
        if num % 2 == 0:
            yield num
```

```python
my_list = [1, 2, 3, 4, 5, 6, 7, 8]
for i in even_numbers(my_list):
    print(i)
```

### **Question 7.  What are the advantages of using generators over regular functions?**

The main advantage of using generators instead of regular functions are as follows:
1. Generators are memory efficient, it only produce values one at a time and on-demand and do not store the entire sequence of values in memory.
whereas, regular functions generate and store all the values in memory at once. This can cause memory issues when dealing with large data sets.
2. Generators use lazy evaluation which means they only compute values when needed.
whereas, Regular functions that return lists or other data structures compute all values upfront, regardless of whether they are all used or not.
3. generators improves the performance when using large datasets, it allow to start processing data immediately instead of waiting for the entire dataset to be computed and stored in memory.
4. Generators make it easier to write complex iterators by simplifying the code. Instead of managing state with loops and counters, the generator function handles state automatically via the yield keyword.
whereas, regular functions may require more lines of code to handle iterating, state management, and returning intermediate results.
5. Generators are ideal for processing streams of data (example, data from network requests, real-time sensor data etc.) because they process one piece of data at a time, without waiting for the entire dataset to be available.
Whereas, with regular functions might need to load all data into memory before processing it.
6. Using generators for tasks like filtering, transforming, and processing data can lead to more clean and readable codes. With a generator we can encapsulate the logic of data generation or transformation without needing to manage state manually.
whereas, Regular functions often involve managing more complex data structures and looping logic, which can make the code harder to follow.

### **Question 9. Explain the purpose and usage of the map() function in Python.**

Purpose and Usage of the **map()** Function in Python is to apply a given function to all items in an iterable (like list, tuple, etc.) and return a map object (which is an iterator) that yields the results. It allows for functional programming by transforming data in a concise and efficient manner.

Syntax of map:

map(function, iterable, ...)

Some important use of map function are:

**1. Transforming data:** without using loops we can apply function to each element of iterable.

```python
#example 1
#Convert a list of strings to integers.
numbers = map(int, ["1", "3", "7"])
print(list(numbers))
```

**2. Data cleaning and processing:** we can Apply a normalization function to each element in a dataset.*italicized text*

```python
names = ["Azlan", "Binod", "Chandan"]
lower_names = map(str.lower, names)
print(list(lower_names))
```

**3. Combining Data:** we can merge multiple iterables.

```python
Key = ["Name", "Age","City"]
Value = ["Azlan", 27, "Patna"]
result = list(map(lambda x, y: (x, y), Key, Value))
print(result)
```

**4. Concise and readable code:**  Map() is a core functional programming concept that helps you write cleaner code by avoiding explicit loops.

```python
numbers = [10, 20, 30, 40]
squares = map(lambda x: x**2, numbers)
print(list(squares))
```

### **Question. 10  What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?**

The map(), reduce(), and filter() functions in Python are all higher-order functions that allow for functional programming operations on iterables.These functions are powerful tools for functional programming in Python and are frequently used for data manipulation, transformation, and filtering tasks. Below is the explained difference between all these function.

1. map() is used when we need to execute a specified function for each of item of an iterable.
2. reduce() is used for folding and reduction of items and combine elements into a single value. To use reduce(), we need to import it from the functools module.
3. filter() is used to filter elements from an iterable based on same condition.


```python
#Example, using map() to get the length of each string.
fruits = ['apple', 'banana', 'watermelon']
lengths = map(len, fruits)
print(list(lengths))
```

```python
#Example, using reduce() to get sum of all numbers
from functools import reduce
numbers = [10, 20, 30, 40, 50]
sum_of_numbers = reduce(lambda x, y: x + y, numbers)
print(sum_of_numbers)
```

```python
#Example, using fliter() to get non negative numbers
numbers = [12, -5, 3, 13, -2, 0, 6, -1]
non_negative_numbers = filter(lambda x: x >= 0, numbers)
print(list(non_negative_numbers))
```
